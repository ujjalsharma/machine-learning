```python
# Implementing simple linear regression on a 'salary vs years of experience' dataset.
```

```python
# Importing required libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Importing dataset
dataset = pd.read_csv('Salary_Data.csv')

# Defining the X and y
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,1].values
```

```python
# Splitting the dataset into Training and Test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)
```

```python
# Fitting the training data to a simple linear regressor
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Predicting the values of Training set and Test set as per the model
y_train_predict = regressor.predict(X_train)
y_test_predict = regressor.predict(X_test)
```

```python
# Visualizing the results of Training set
plt.scatter(X_train, y_train, color='red', label='Training Data')
plt.plot(X_train, y_train_predict, color='blue', label='Linear model fit')
plt.title('Salary vs Experience (Training set)')
plt.ylabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
# Visualizing the results of Test set
plt.scatter(X_test, y_test, color='red', label='Test Data')
plt.plot(X_train, y_train_predict, color='blue', label='Linear model fit')
plt.title('Salary vs Experience (Test set)')
plt.ylabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
