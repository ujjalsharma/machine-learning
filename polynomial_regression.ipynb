```python
# Importing the required libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Importing the dataset
dataset = pd.read_csv('Position_Salaries.csv')

# Defining X and y
X = dataset.iloc[:,1:2].values
y = dataset.iloc[:,2].values
```

```python
# Fitting the data using linear regression model
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
# Visualising the Linear Regression model fit
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg.predict(X), color='blue')
plt.title('Linear Regression Model Fit')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
# Fitting the data using polynomial regression of degree 2
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X)
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly, y)
```

```python
# Visualizing the fit using polynomial regression of degree 2
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg2.predict(X_poly), color='blue')
plt.title("Degree 2 Polynomial Regression Fit")
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
# Fitting the data using polynomial regression of degree 3
poly_reg = PolynomialFeatures(degree=3)
X_poly = poly_reg.fit_transform(X)
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly, y)
```

```python
# Visualizing the fit using polynomial regression of degree 3
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg2.predict(X_poly), color='blue')
plt.title("Degree 3 Polynomial Regression Fit")
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
# Fitting the data using polynomial regression of degree 4
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly, y)
```

```python
# Visualizing the fit using polynomial regression of degree 4
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg2.predict(X_poly), color='blue')
plt.title("Degree 4 Polynomial Regression Fit")
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
# Predicting a new result with Linear Regression
lin_reg.predict([[6.5]])
```

```python
# Predicting a new result with Polynomial Regression
lin_reg2.predict(poly_reg.fit_transform([[6.5]]))
```
